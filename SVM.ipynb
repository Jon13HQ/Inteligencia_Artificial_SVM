```python
#Importamos la bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn import svm
from sklearn.datasets import make_blobs
import seaborn as sns
from pandas import read_csv
import io
```

```python
#Esta parte es para poder cargar los datos
from google.colab import files

features = files.upload()
respuesta = files.upload()
```

```python
#Nombramos las columnas y a las etiquetas que son el archivo de respuestas
features_name=['A','B','C','D','E','F','G','H','I']
target_names=['0','1']

#Creamos una variable y le Asignamos los datos del features
features_1 = read_csv(io.BytesIO(features['features_A.csv']), header=None, names=features_name)
data_features = features_1.values
print(data_features)


#Creamos unas variable y le Asignamos los datos de la respuesta
respuesta_1 = read_csv(io.BytesIO(respuesta['respuesta.csv']), header=None, names=['Objetivo'])
data_respuesta = respuesta_1.values
```

```python
for i in range (30):
  #una vez que se dividieron los datos, dividimos los datos en entrenamiento y prueba con el método train_test_split
  # de la libreria model_selection en Scikit-learn.
  X_train, X_test, y_train, y_test = train_test_split(data_features, data_respuesta)

  #Usamos la biblioteca Scikit-Learn
  #Entrenamos el algoritmo, El entrenamiento de datos se realiza utilizando la biblioteca SVM.
  #usamos una biblioteca para la clasificación. Esta biblioteca es SVC esta admite la clase de clasificador de vectores 
  #Tenemos que especificar el tipo de kernel que estamos usando en esta clase, en este caso se usa linear
  from sklearn.svm import SVC
  svclassifier = SVC(kernel='linear')
  svclassifier.fit(X_train, y_train)

  #Despues del entrenamiento usamos este metodo para realizar las predicciones
  y_pred = svclassifier.predict(X_test)

  #Y por ultimo  con esto se evaluara el algoritmo y nos mostrara el resultado del svm
  from sklearn.metrics import classification_report, confusion_matrix
  print(confusion_matrix(y_test,y_pred))
  print(classification_report(y_test,y_pred))

  #Obtener vectores de Soporte
  support_vectors = svclassifier.support_vectors_

  # Visualize support vectors
  plt.scatter(X_train[:,0], X_train[:,1])
  plt.scatter(support_vectors[:,0], support_vectors[:,1], color='red')
  plt.title('Linearly separable data with support vectors')
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.show()
print (i)
```
